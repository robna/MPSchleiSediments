```python
import pandas as pd
import numpy as np
import altair as alt
```

## Compare model ensemble compositions

```python
ens_comp = pd.read_csv('../data/ensemble_compositions_20230918.csv', index_col=0).reset_index().melt(id_vars=['FOLDS', 'model type'], var_name='scorer', value_name='composition')
ens_comp
```

```python
# make an altair chart of the ensemble compositions: y-axis is the composition normalized to 1 displayer as percentage stacked areas, x-axis is the FOLDS, color is the model type in the order ['GLM', 'XGBlinear', 'XGBtree', 'RF'], column is the scorer
alt.Chart(ens_comp).mark_area().encode(
    x='FOLDS:O',
    y=alt.Y('composition:Q', stack='normalize'),
    color='model type:N',
    column='scorer:N'
)
```
